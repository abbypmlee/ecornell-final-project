```python
# YOUR CODE HERE
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
# Isolate features; drop label and columns that are irrelevant or result in feature leakage
# Include 'country' for now to group by later
features = df.drop(columns=
                   ['Life Ladder', 
                    'year', 
                    'GINI index (World Bank estimate)', 
                    'Standard deviation of ladder by country-year', 
                    'Standard deviation/Mean of ladder by country-year']).columns
```

```python
# Look at one row that contains a missing value for GINI index
print("Row 0:  " + str(df['GINI index (World Bank estimate), average 2000-15'][0]))

# Compute mean for all non null age values
for col in nan_cols:
    mean=df[col].mean()

    # Fill all missing values with the mean
    df[col].fillna(value=mean, inplace=True)

# Look at one of the rows that contained a missing value for GINI index. 
# It should now contain the mean
print("Row 0:  " + str(df['GINI index (World Bank estimate), average 2000-15'][0]))
```

```python
# Checking if all NaNs are filled
np.any(np.sum(df.isnull(), axis=0))
```

```python
# Plot the distributions of features
fig, axes = plt.subplots(3, 5, figsize=(20, 15)) # 3 rows, 5 columns
axes = axes.flatten()

for i, col in enumerate(df[features].select_dtypes(include=np.number)):
    sns.histplot(x=col, data=df, ax=axes[i])

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show()
```

```python
# Winsorize outliers of features
for col in df[features].select_dtypes(include=np.number):
    df[col] = stats.mstats.winsorize(df[col], limits=[0.01, 0.01])
```

```python
# Plot the distributions of features again 
fig, axes = plt.subplots(3, 5, figsize=(20, 15)) # 3 rows, 5 columns
axes = axes.flatten()

for i, col in enumerate(df[features].select_dtypes(include=np.number)):
    sns.histplot(x=col, data=df, ax=axes[i])

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show()
```

```python
# Split data into features and labels
X = df[features].groupby(['country']).mean()
y = df.groupby(['country']).mean()['Life Ladder']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Scale the feature data
scaler = StandardScaler()

# Fit on the training data
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), index=X_train.index)

# Just transform on the test data to avoid feataure leakage
X_test_scaled = pd.DataFrame(scaler.transform(X_test))
```

```python
# Create and train model    
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions on scaled data
y_pred = model.predict(X_test_scaled)
```

```python
scores = cross_val_score(model, X, y, cv=5, scoring='r2')
print('R^2 scores for the five training/test iterations on the validation sets:')
for score in scores:
    print(score) 
```

```python
# RMSE
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, y_pred)))

# R^2
print(' R^2 =   %.2f'
      % r2_score(y_test, y_pred))
```

```python
feature_importances = model.coef_
plt.barh(X.columns, feature_importances)
plt.title("Feature Coefficients")
plt.show()
```

## Alternative model: Gradient Boosted Decision Tree

Even though linear regression appeared to perform well on my data, I also wanted to try a method that was new to me, as I was already familiar with linear regression from my classes. As a result, I decided to practice model selection with a Gradient Boosted Decision Tree.

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import GridSearchCV
```

```python
# Testing two GBDT models
# Max depth of 2
gbdt_2_model = GradientBoostingRegressor(n_estimators = 50, max_depth = 2)
gbdt_2_model.fit(X_train_scaled, y_train)
gbdt_2_predictions = gbdt_2_model.predict(X_test_scaled)

# Max depth of 10
gbdt_10_model = GradientBoostingRegressor(n_estimators = 50, max_depth = 10)
gbdt_10_model.fit(X_train_scaled, y_train)
gbdt_10_predictions = gbdt_10_model.predict(X_test_scaled)
```

```python
# Print metrics for model 1, max depth = 2
# RMSE
print('\nModel Performance (max_depth=2)\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, gbdt_2_predictions)))

# R^2
print(' R^2 =   %.2f'
      % r2_score(y_test, gbdt_2_predictions))
```

```python
# Print metrics for model 1, max depth = 2
# RMSE
print('\nModel Performance (max_depth=10)\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, gbdt_10_predictions)))

# R^2
print(' R^2 =   %.2f'
      % r2_score(y_test, gbdt_10_predictions))
```

```python
# Hyperparameter values for 'max_depth'
hyperparams_depth = [2**n for n in range(2,6)]

# Hyperparameter values for 'min_samples_leaf'
hyperparams_leaf = [25*2**n for n in range(0,4)]

# Hyperparameter values for 'n_estimators'
hyperparams_estimators = [50*2**n for n in range(0,4)]

# Create parameter grid
param_grid={'max_depth':hyperparams_depth, 'min_samples_leaf':hyperparams_leaf, 'n_estimators':hyperparams_estimators}
param_grid
```

```python
print('Running Grid Search...')
gbdt_model = GradientBoostingRegressor()

# 2. Run a Grid Search with 5-fold cross-validation
grid = GridSearchCV(gbdt_model, param_grid, cv=5)
grid_search = grid.fit(X_train_scaled, y_train)
print('Done')
```

```python
# Print best-performing hyperparameter configuration
best_params = grid_search.best_params_
print('Optimal hyperparameters: {0}'.format(best_params))

# print best accuracy score resulting from this configuration
print('mean R^2 score across 5 folds: {0}'.format(grid_search.best_score_))
```

```python
gbdt_predictions = grid_search.best_estimator_.predict(X_test_scaled)

# The RMSE
print('\nOptimal GBDT Model Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, gbdt_predictions)))

# R^2
print(' R^2 =   %.2f'
      % r2_score(y_test, gbdt_predictions))
```

```python
gbdt_feature_importances = grid_search.best_estimator_.feature_importances_
plt.barh(X.columns, gbdt_feature_importances)
plt.title("Feature Importances")
plt.show()
```
